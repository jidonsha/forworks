1 購物車
某電子商務網站需要實現一個虛擬的購物車，以便讓它的客戶能夠同時購買多件商品，從而獲
得更好的購物體驗。在客戶訪問該網站的過程中，他們可以一邊流覽商品，一邊將自己滿意的商品
添加到購物車中，等到選購完畢，再一次性結算購物車中的所有商品。購物車結算過程涉及到一些
業務邏輯，需要撰寫程式自動計算最後購買者需要支付的實際金額。
其中一個影響結算金額的因素是折扣。網站會在一些節日或者
期間 推出某 類 商品的折扣，例如
中元節的時候，會有針對食品的折扣，而在中秋節的時候，會有針對日用品的折扣，在計算購物車
結算金額的時候，需 要根據結算當天的折扣計算出折扣後的金額。
另外，網站為了促銷，曾經在過去一年內派發了一些優惠券，這些優惠券只要有效期內，都能
夠用在任何一次購物中，但是每次結算只能用一張。這些優惠通常是 滿 1000 折 200” 這樣的形
式，表示總金額如果超過 1000 則立即折扣 200 。
簡化掉系統的其他部分，你需要撰寫一個程式，根據輸入的購物車資訊和促銷資訊來計算使用
者需要支付的結算金額。我們初步支援以下的產品目錄：
• 電子：ipad，iphone，螢幕，筆記型電腦，鍵盤
• 食品：麵包，餅乾，蛋糕，牛肉，魚，蔬菜
• 日用品：餐巾紙，收納箱，咖啡杯，雨傘
• 酒類：啤酒，白酒，伏特加
依據以上描述完成您的主要程式，以下是兩個測試情境，用以證明你的設計能確實執行，請將
這兩個情境以適當的資料型態儲存，並能載入你的程式中，計算出正確結果，再將執行結果截圖。
當然，即使任意修改情境中的數值、日期或折扣項目，程式也必須正確運算，你可以另外附上其它
單元測試來證明之。

```python
from datetime import datetime, timedelta, time
from random import randrange
import numpy as np
import time
import datetime
```

====== 情境一 ========

結帳日
期 201 8 .11.11
促銷
活動：
1. 2018.11.11 當天 電子 類商品一律 7 折
2. 2018.11. 0 1 2018.11. 1 5 酒 類商品一律 8 折
購物車
內容
1. ipad: 2399.00 * 1
2. 螢幕 1799.00 * 1
3. 啤酒 25.00 * 12
4. 麵包 9.00 * 5

使用優惠券 滿 1000 折 200 2019.3.2 到期

```python
ccc_list = [['ipad',2399.00], ['iphone',23990.00], ['螢幕',1799.00], ['筆記型電腦',19990.00], ['鍵盤',499.00]]
food_list = [['麵包',9.00], ['餅乾',5.00],['蛋糕',199.00], ['牛肉',99.00], ['魚',88.00], ['蔬菜',5.98]]
tool_list = [['餐巾紙',3.20], ['收納箱',399.00], ['咖啡杯',33.00], ['雨傘',699.00]]
wine_list = [['啤酒',25.00], ['白酒',199.00], ['伏特加',299.00]]
total_lists = ccc_list+food_list+tool_list+wine_list

money = input("請輸入您擁有的金額： ")
if money.isdigit(): # 如果輸入的資料是一個數字
    money = int(money)
    lists = '---------------總商品列表----------------\n編號    商品    單價(NTD) \n'

    for index,item in enumerate(total_lists): # 列舉出全部類別列表的資料
#            if item in ccc_list:
#                print(item[0])
            lists += "No.%d   %s   %d" %(index+1,item[0],item[1]) + "\n"
print(lists)

someday = datetime.date(2018,11,11)
specialday = datetime.date(2018,11,11)
Start = datetime.date(2018,11,1)
End = datetime.date(2018,11,15)
cart = []
counts = 0
amount = 1  #購買數量
amounts = 1 #產品 x 數量
coupon = 200 #優惠券：滿1000折200
deadline = datetime.date(2019,3,2) #優惠券：2019.3.2到期
num = input("請輸入要購買的 商品編號（按回車即購買，輸入 q 退出）：")
amount = input("請輸入要購買的 數量（按回車即購買，輸入 q 退出）：")
while num and amount != 'q' :
    if num.isdigit() and amount.isdigit():
        if(int(num) <= len(total_lists)):
            num = int(num) - 1
            item = total_lists[int(num)]
            if money >= item[1]:
                cart.append(item)
                if (item in ccc_list and someday == specialday):
                    print("您購買[" + item[0] + "]電子類商品一律 7 折，原價值[" + str(item[1]) + "]" + "，數量:"+ str(amount))
                    costoff = item[1] * 0.7
#                       print(item[0],int(item[1]))
                    print("折扣後價值[" + str(costoff) + "]")
                elif (item in wine_list and someday >= Start and someday <= End):
                    print("您購買[" + item[0] + "]酒類商品一律 8 折，原價值[" + str(item[1]) + "]" + "，數量:"+ str(amount))
                    costoff = item[1] * 0.8
#                       print(item[0],int(item[1]))
                    print("折扣後價值[" + str(costoff) + "]")
                else :
                    print("您已購買[" + item[0] + "]，價值[" + str(item[1]) + "，數量:"+ str(amount) +"]，您的餘額：" + str(money))
                amounts = int(item[1]) * int(amount)
                money -= amounts
                counts += amounts
            else:
                print("餘額不足...請選擇請他商品")
        else:
            print("商品編號不存在")
    else:
        print("您輸入的編號好像不對哦...")
    num = input("請輸入要購買的 商品編號（按回車即購買，輸入 q 退出）：")
    amount = input("請輸入要購買的 數量（按回車即購買，輸入 q 退出）：")
        
conclusion = "--------------購物車列表-------------\n編號    商品    單價(NTD)\n"
for index,item in enumerate(cart):
    conclusion += "No.%d, %s, %d" %(index+1,item[0],item[1]) + "\n"
conclusion += "消費總金額(使用優惠券前): " + str(counts) + "\n" + "帳戶餘額: "+ str(money) + "\n"
if (someday <= deadline and counts >= 1000):
    answer = input("您有優惠券滿1000 折200，2019.3.2 到期（請問是否使用優惠券，輸入 Y or y 折扣）：")
    if (answer == 'Y' or 'y'):
        counts = counts - coupon
        coupon = 0
        conclusion += "消費總金額(使用優惠券後): " + str(counts) + "\n" + "帳戶餘額: "+ str(money)
    else :
        counts = counts
        conclusion += "消費總金額(不使用優惠券): " + str(counts) + "\n" + "帳戶餘額: "+ str(money) 
print(conclusion)
```
